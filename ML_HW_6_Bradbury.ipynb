```python
import pandas as pd
import numpy as np
df = pd.read_csv('https://raw.githubusercontent.com/lmanikon/lmanikon.github.io/master/teaching/datasets/KaggleAdmissions.csv')
df.head()
```

### Q1 [18 points]

#### 1. Create a new dataframe `df_sub` by dropping `Serial No.` these feature columns in `df` : 

#### 2. Standardize only these attributes in `df_sub` using the function `StandardScaler()`: 
- `GRE Score`, `TOEFL Score`

#### 3. Perform normalization only on these attributes in `df_sub` using the function `MinMaxScaler()`
- `University Rating`, `SOP`, `LOR`, `CGPA`, `Research` 
    
#### Note that after steps 3 and 4, make sure you still have the transformed values saved in `df_sub` 

#### 4. Create a new column named `Admit` using original column `Chance of Admit ` to create a discrete set of class labels using these conditions. Then drop `Chance of Admit ` column from `df_sub`. 
- Convert to 2, if `Chance of Admit ` value is `>= 0.73`
- 1, if `Chance of Admit ` value is `< 0.73`



```python
#Answering the above questions in the same order as listed

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler

df_sub = df.drop(columns = ['Serial No.'])

scaler = StandardScaler()
cols_to_scale = ['GRE Score','TOEFL Score']
scaler.fit(df_sub[cols_to_scale])
df_sub[cols_to_scale] = scaler.transform(df_sub[cols_to_scale])
df_sub.head()

mms = MinMaxScaler()
cols_to_norm = ['University Rating', 'SOP', 'LOR ', 'CGPA', 'Research']
mms.fit(df_sub[cols_to_norm])
df_sub[cols_to_norm] = mms.transform(df_sub[cols_to_norm])

df_sub['Admit'] = np.where(df_sub['Chance of Admit '] < 0.73, 1, 2)
df_sub = df_sub.drop(columns = ['Chance of Admit '])
df_sub.rename(columns = {'LOR ':'LOR'}, inplace = True)

df_sub.head(10)
```

### Q2 [15 points]
#### 1. Split the data into `X` and `y` for feature columns and class label column respectively. 
- Feature columns (X): `GRE Score`, `TOEFL Score`, `University Rating`, `SOP`, `LOR`, `CGPA`, `Research` 
- CLass label column (y): `Admit`

#### 2. Using `X` and `y` variables representing features and class labels, perform train_test_split operation to build training (`X_train`, `y_train`) and testing data (`X_test`, `y_test`). 
- Use test_size=0.3, random_state=42 as the parameters for train_test_split() function. 

#### 3. Train the randomforest classifier (initialized as variable `rf`) using these parameters: max_depth=4, random_state=0. 
- Using the trained model `rf` to first compute accuracy score and assign it to variable `acc_rf`. 
- Then compute the impurity-based feature importances.
- Append the names of these top-3 features to a list `impFeatures`. Please make sure you type the feature names exactly as in `df_sub` 

#### 4. Train a K-Nearest Neighbors classifier (initialized as variable `knn1` and `knn2`) using these parameters: n_neighbors=4 and n_neighbors=20.
- Using the trained models `knn1` and `knn2` on training data, compute the accuracy score using test data and assign that to variables `acc_knn_4` and `acc_knn_20` respectively for `k=4` and `k=20`. 

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

#Answering the above questions in the same order as listed

impFeatures=[] # Important Features

# Data Split
X = df_sub[['GRE Score', 'TOEFL Score', 'University Rating', 'SOP', 'LOR', 'CGPA', 'Research']]
y = df_sub['Admit']

# Train/Test
X_train, X_test, y_train, y_test= train_test_split(X, y, test_size = 0.3, random_state =42)

# Random Forest
rf = RandomForestClassifier(max_depth = 4, random_state = 0)
rf.fit(X_train, y_train)
y_pred_rf = rf.predict(X_test)
acc_rf = accuracy_score(y_test, y_pred_rf)
print(round(acc_rf,2)) # Accuracy

features = pd.DataFrame({'features':rf.feature_names_in_, 'importance':rf.feature_importances_})
features = features.sort_values(by='importance', ascending=False)
for x in range(3): # Top 3 Features
  impFeatures.append(features.iloc[x,0])
print(impFeatures)

# kNN
knn1 = KNeighborsClassifier(n_neighbors = 4)
knn1.fit(X_train, y_train)
y_pred_knn = knn1.predict(X_test)
acc_knn_4 = accuracy_score(y_test, y_pred_knn)
print(round(acc_knn_4,2))

knn2 = KNeighborsClassifier(n_neighbors = 20)
knn2.fit(X_train, y_train)
y_pred_knn = knn2.predict(X_test)
acc_knn_20 = accuracy_score(y_test, y_pred_knn)
print(round(acc_knn_20,2))
```

```python
#[8 points] Test cell-4
#DO NOT MODIFY/DELETE THIS CELL 
assert (rf.max_depth)==4
assert (rf.n_estimators)==100
assert round(acc_rf,2)==0.9
assert set(impFeatures)=={'GRE Score', 'CGPA', 'TOEFL Score'}
```

```python
#[6 points] Test cell-5
#DO NOT MODIFY/DELETE THIS CELL 
assert knn1.metric=="minkowski"
assert knn2.metric=="minkowski"
assert (round(acc_knn_4,2))==0.87
assert (round(acc_knn_20, 2))==0.9
```
