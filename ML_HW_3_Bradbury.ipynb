#### Q5 [10 points]. Binarization and Discretization. Please answer these questions in the following order. 
<ol>
    <li>Consider the dataframe `df` (after all the manipulations in Q1) and convert all values of feature `class` as follows
        <ol>
            <li>if value is 'First' convert it to 1</li>
            <li>if the value is not 'First' and anything else, convert it to 0</li>
        </ol>
    </li>
    <li>Consider the column `fare` and convert the continuous values to discrete values using these instructions: 
        <ol>
            <li>if value is less than (`<`) 31, convert it to 1</li>
            <li>if the value greater than or equal to 31, convert it to 2</li>
        </ol>
    </li>
</ol>
- Note that by performing these two operations, feature columns `fare` is discretized and `class` is binarized. 

```python
df['class'] = [1 if x == 'First' else 0 for x in df['class']]

df.loc[df['fare'] < 31, 'fare'] = 1
df.loc[df['fare'] >= 31, 'fare'] = 2

df.head()
```
